```python
#@title Set up tpattern — run once (code hidden) { display-mode: "form" }
# ----------------------------------------------------------------------------
# NOTE FOR THE CURIOUS READER
# The tpattern source is bundled (gzipped + base64) into this one cell purely so
# the notebook is self-contained and runs in Google Colab with nothing to
# install or download. It is hidden only for convenience — nothing here is
# closed. The full, readable, commented Python source is openly available:
#   archive (permanent DOI): https://doi.org/10.5281/zenodo.21397543
#   source repository:       https://github.com/ajcallaway/TPattern
# The cell below simply unpacks that same source to a local 'tpattern/' folder.
# ----------------------------------------------------------------------------
import base64, gzip, json, os
FILES = json.loads(gzip.decompress(base64.b64decode('''''')).decode())
os.makedirs('tpattern', exist_ok=True)
for name, src in FILES.items(): open('tpattern/'+name, 'w').write(src)
print('tpattern ready:', len(FILES), 'modules')
```
